# Kompozycja z drzew binarnych

Kompozycja składa się z 4 drzew binarnych. Każde następne wywołanie kodu generuje frakatal prostopadły do poprzednika. Mamy identyczne odbicia lustrzane pojedynczego fraktala drzewa binarnego w poziomie i w pionie.

Taki zabieg pozwala nam uzyskać kompozycję, z której możemy utworzyć coś na kształt koniczyny.

## Właściwości

- **Symetryczność**: Symetryczność w poziomie i w pionie oraz wględem środka.

```python
import matplotlib.pyplot as plt
import math

def draw_tree(ax, x, y, angle, length, level):
    if level == 0:
        return
    else:
        x_new = x + length * math.cos(math.radians(angle))
        y_new = y + length * math.sin(math.radians(angle))
        ax.plot([x, x_new], [y, y_new], 'k-', lw=1)


        draw_tree(ax, x_new, y_new, angle - 45, length * 0.6, level - 1)
        draw_tree(ax, x_new, y_new, angle + 45, length * 0.6, level - 1)   
            
fig, ax = plt.subplots()
ax.axis('equal')
ax.axis('off')


# Rysowanie czterech kolejno prostopadadłych drzew binarnych
draw_tree(ax, 0, 0, 180, 5, 5)
draw_tree(ax, 0, 0, 90, 5, 5)
draw_tree(ax, 0, 0, 0, 5, 5)
draw_tree(ax, 0, 0, -90, 5, 5)

plt.show()
```

# Kompozycja z drzew binarnych - Wariant losowy

Można oczywiście również zmieniać kąty, aby fraktale pojedynczych drzew binarnych nie były do siebie prostopadłe. Każdy początkowy fraktal jest generowany pod losowym kątem z kolejnej ćwiartki układu współrzędnych. Tracimy wtedy symetrzyczność.

## Właściwości

- **Losowość**: Za każdym razem generowana jest inna kompozycja fraktali.

```python
import matplotlib.pyplot as plt
import random
import math

def draw_tree(ax, x, y, angle, length, level):
    if level == 0:
        return
    else:
        x_new = x + length * math.cos(math.radians(angle))
        y_new = y + length * math.sin(math.radians(angle))
        ax.plot([x, x_new], [y, y_new], 'k-', lw=1)


        draw_tree(ax, x_new, y_new, angle - 45, length * 0.6, level - 1)
        draw_tree(ax, x_new, y_new, angle + 45, length * 0.6, level - 1)   
            
fig, ax = plt.subplots()
ax.axis('equal')
ax.axis('off')


# Rysowanie czterech drzew binarnych w różnych kierunkach
draw_tree(ax, 0, 0, random.randint(90, 180), 5, 5)
draw_tree(ax, 0, 0, random.randint(0, 90), 5, 5)
draw_tree(ax, 0, 0, - random.randint(90, 180), 5, 5)
draw_tree(ax, 0, 0, - random.randint(0, 90), 5, 5)

plt.show()
```
